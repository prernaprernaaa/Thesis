```python
import numpy as np
import matplotlib.pyplot as plt
import random

def func(x, y, alpha):
    if x <= 0:
        f1 = (alpha / (1 - x)) + y
        return f1
    
    elif 0< x < alpha + y:
        f2 = alpha + y
        return f2
    
    else:  # x >= alpha + y 
        f3 = -1
        return f3


def rulkov(xn, yn, T , alpha, mu, sigma):

    xlist = [xn]
    ylist = [yn]

    for i in range(1,T):
        xnew = func(xn, yn, alpha)
        xlist.append(xnew)
        ynew = yn - mu*(xn +1) + mu*sigma
        ylist.append(ynew)
        xn = xnew
        yn = ynew
    return xlist, ylist   


```

```python
#parameters
T = 3000
alpha = 4
mu = 0.001
sigma =  0.01

#fig2 phase plot

xlist, ylist = rulkov(-1, -4.93, 4000, 4, 0.001, 0.01)
time = np.arange(4000)
plt.plot(xlist[2000:],ylist[2000:])
 
```

```python
#fig2


# Parameters
y = -3.93
alpha = 6
x0_values = np.linspace(-5, 5, 500)

# Compute xn and xn+1
xn = x0_values
xn1 = np.array([func(x, y, alpha) for x in xn])

# Plot
plt.figure(figsize=(10, 6))
plt.plot(xn, xn1)

plt.title("x_n vs x_n+1")
```

```python

plt.figure(figsize=(12, 6))
xsil, ysil = rulkov(0.7887, 0.4678, 6000, 4, 0.001, -0.05)
time = np.arange(6000)

# print (xlist, ylist)

plt.figure(figsize=(10, 6))
plt.title('Silent Neuron', fontsize=16, pad = 7)
plt.ylabel('Membrane Potential (x)' , fontsize = 16,  labelpad=15)
plt.xlabel('Time(n)' , fontsize = 16,  labelpad=15)

plt.plot(time[2000:],xsil[2000:], linewidth = 1)
```

```python

#font settings
plt.rc('text', usetex=True)
plt.rc('font', family='serif')


T = 13000
l = 10000

time = np.arange(T-l)
xlis2a, ylis2a = rulkov(-1, 0.3, T, 4, 0.001, 0.05)


plt.figure(figsize=(10, 6))
plt.title('Spiking Neuron', fontsize=16, pad = 7)
plt.plot(time,xlis2a[l:])
plt.ylabel('Membrane Potential (x)' , fontsize = 16,  labelpad=15)
plt.xlabel('Time(n)' , fontsize = 16,  labelpad=15)
plt.show()
```
